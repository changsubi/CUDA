```python
%%writefile test1.c
#include <stdio.h>
#include <stdlib.h>
#include <unistd.h>
#include <pthread.h>

#define N 10000

int g;
void* func(void* data){
	++g;
}

int main()
{
	pthread_t pthread[N];
	for (int i = 0; i < N; ++i)
	{
		pthread_create(&pthread[i], NULL, func, NULL);
	}

	for (int i = 0; i < N; ++i)
	{
		pthread_join(pthread[i], NULL);
	}

	printf("%d\n",g);
	return 0;
}
```
